```python
import numpy as np
import pandas as pd

data = pd.read_csv(r'C:\Users\Kelvin Yu\Documents\Sophomore Year\COS 424\Final Project\data\data.csv', low_memory = False)
```

```python
# drop link, website, Expansion Plans, Founder First Name, Founder Last Name, Founder LinkedIn, Founder Twitter, Founder AngelList
print("Before shape", data.shape)

data = data.drop(['LINK', 'Website', 'Expansion Plans', 'Founder First Name', 'Founder Last Name', 'Founder LinkedIn', 'Founder Twitter', 'Founder AngelList'], axis=1)

print("After shape", data.shape)
```

```python
# drop rows with "Week to Date" and "Month to Date" values...not relevant to our study
data = data[data['COMPANY'] != 'Week to Date']
data = data[data['COMPANY'] != 'Month to Date']
data = data[data['COMPANY'] != 'Total']
data = data.dropna(thresh=5)
print("New shape", data.shape)
```

```python
# create new columns, one for year and one for month
year = pd.Series(data='DATE RAISE ANNOUNCED')
month = pd.Series(data='DATE RAISE ANNOUNCED')

# iterate through rows
for index, row in data.iterrows():
    # if DATE RAISE ANNOUNCED contains "Apr-" add april to month column, etc.

    if ('Apr' in row['DATE RAISE ANNOUNCED']):
        month.at[index] = 4
        
    elif ('Mar' in row['DATE RAISE ANNOUNCED']):
        month.at[index] = 3

    elif ('Feb' in row['DATE RAISE ANNOUNCED']):
        month.at[index] = 2
            
    elif ('Jan' in row['DATE RAISE ANNOUNCED']):
        month.at[index] = 1
        
    elif ('Dec' in row['DATE RAISE ANNOUNCED']):
        month.at[index] = 12
        
    elif ('Nov' in row['DATE RAISE ANNOUNCED']):
        month.at[index] = 11
    
    elif ('Oct' in row['DATE RAISE ANNOUNCED']):
        month.at[index] = 10
        
    elif ('Sep' in row['DATE RAISE ANNOUNCED']):
        month.at[index] = 9
        
    elif ('Aug' in row['DATE RAISE ANNOUNCED']):
        month.at[index] = 8

    elif ('Jul' in row['DATE RAISE ANNOUNCED']):
        month.at[index] = 7
        
    elif ('Jun' in row['DATE RAISE ANNOUNCED']):
        month.at[index] = 6
    
    elif ('May' in row['DATE RAISE ANNOUNCED']):
        month.at[index] = 5

# if DATE RAISE ANNOUNCED contains "16" add 0 to year column, "-17" = 1, etc.
for index, row in data.iterrows():
    
    if ('16' in row['DATE RAISE ANNOUNCED']):
        year.at[index] = 0
        
    elif ('17' in row['DATE RAISE ANNOUNCED']):
        year.at[index] = 1

    elif ('18' in row['DATE RAISE ANNOUNCED']):
        year.at[index] = 2
            
    elif ('19' in row['DATE RAISE ANNOUNCED']):
        year.at[index] = 3
        
# drop DATE RAISE ANNOUNCED feature
data = data.drop(['DATE RAISE ANNOUNCED'], axis=1)

# add year and month as features to pandas
data['Year'] = year
data['Month'] = month
```

```python
# drop rows with unknown investors
print("Before: ", data.shape)
data = data[data['TOP INVESTORS (in this round)'] != 'U']
print("After: ", data.shape)
```

```python
# add city and region features and drop location feature

location = data['HQ Location']
location = location.str.split(pat=',', n=-1, expand=True)
data['City'] = location[0]
data['Region'] = location[1]

data = data.drop(['HQ Location'], axis=1)
```

```python
# do the same as above except with investors
first_investor = pd.Series(data='Year')
second_investor = pd.Series(data='Year')
third_investor = pd.Series(data='Year')

investors = data['TOP INVESTORS (in this round)'].str.split(pat=',', n=-1, expand=True)
# top investor only
investors = investors.dropna(axis=1, thresh = 5000)

data['First Investor'] = investors[0].fillna(0)

data = data.drop('TOP INVESTORS (in this round)', axis=1)
```

```python
# clean categories (mispellings, some mean the same thing, etc.)
category = pd.Series(data = 'Category')

# first, turn all words into lowercase
for index, row in data.iterrows():
    category.at[index] = row['Category'].lower()

data['Category'] = category
```

```python
for index, row in data.iterrows():
    if ('bio' in row['Category'] or 'health' in row['Category']):
        category.at[index] = 'health'
        
    elif ('insur' in row['Category']):
        category.at[index] = 'insurance'
        
    elif ('indus' in row['Category']):
        category.at[index] = 'industrial'
        
    elif ('edtec' in row['Category'] or 'educati' in row['Category']):
        category.at[index] = 'education'
        
    elif ('ytics' in row['Category'] or 'data' in row['Category']):
        category.at[index] = 'analytics'

    elif ('dev' in row['Category']):
        category.at[index] = 'dev'

    elif ('space' in row['Category']):
        category.at[index] = 'space'

    elif ('lear' in row['Category'] or 'ai' in row['Category']):
        category.at[index] = 'AI'

    elif ('drone' in row['Category']):
        category.at[index] = 'drones'
        
    elif ('robot' in row['Category']):
        category.at[index] = 'robotics'

    elif ('foo' in row['Category']):
        category.at[index] = 'food'
        
    elif ('real' in row['Category']):
        category.at[index] = 'real estate'
        
    elif ('social' in row['Category']):
        category.at[index] = 'social'
    
    elif ('cyb' in row['Category']):
        category.at[index] = 'cybersecurity'
        
data['Category'] = category
    
data
```
